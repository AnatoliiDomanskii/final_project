```
data_device_stream = spark.readStream.schema(
    df.schema
).option(
    "maxFilesPerTrigger", 1
).json(
    "/databricks-datasets/iot-stream/data-device/"
)
```

```
user_device_stream_out = data_device_stream.writeStream \
    .format('memory') \
    .outputMode('append') \
    .queryName('q1') \
.start()
```
